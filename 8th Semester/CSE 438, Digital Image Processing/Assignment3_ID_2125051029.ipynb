```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img1 = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.rectangle(img1, (150, 150), (250, 250), (0, 255, 0), -1)
cv2_imshow(img1)
```

```python
img2 = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.circle(img2, (200, 200), 60, (255, 0, 255), -1)
cv2_imshow(img2)
```

```python
bitwise_and = cv2.bitwise_and(img1, img2)
bitwise_or = cv2.bitwise_or(img1, img2)
bitwise_xor = cv2.bitwise_xor(img1, img2)
bitwise_not1 = cv2.bitwise_not(img1)
bitwise_not2 = cv2.bitwise_not(img2)
multiply = cv2.multiply(img1, img2)
alpha_blend = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
subtraction = cv2.subtract(img1, img2)
```

```python
def show_images(images, titles):
    plt.figure(figsize=(15, 10))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(3, 3, i+1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(title)
        plt.axis("off")
    plt.tight_layout()
    plt.show()

show_images(
    [img1, img2, bitwise_and, bitwise_or, bitwise_xor,
     bitwise_not1, bitwise_not2, multiply, alpha_blend],
    ["Green Rectangle", "Magenta Circle", "Bitwise AND", "Bitwise OR", "Bitwise XOR",
     "Bitwise NOT (Rectangle)", "Bitwise NOT (Circle)", "Element-wise Multiply", "Alpha Blending"]
)

plt.figure(figsize=(4, 4))
plt.imshow(cv2.cvtColor(subtraction, cv2.COLOR_BGR2RGB))
plt.title("Image Subtraction")
plt.axis("off")
plt.show()
```
